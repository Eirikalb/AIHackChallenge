This dataset is originated from the US census governments webpages. The data was originially in a Geodatabase format, which is a standard format when working with geographically sensitive data like this dataset. The data gathered describes 23000 block groups in California. A block is described by the US census bureau as "A census block is the smallest geographic unit used by the United States Census Bureau for tabulation of 100-percent data (data collected from all houses, rather than a sample of houses)" they usually have a population of around 1000 people.

The dataset is a result of a 10 question questionaire that every single American citizen has filled out. The answers have then been restructured into 7730 different variables describing the various socioeconomic traits block group in America. We are currently using one subset of the dataset, namely "X15_EDUCATIONAL_ATTAINMENT" for predicting the "PER CAPITA INCOME IN THE PAST 12 MONTHS (IN 2016 INFLATION-ADJUSTED DOLLARS)". Just to clarify each block group originally has 7730 variables, but we only use the 350 or so of them that describe the educational attainment just to be able to actually have some sense of what information is going in to our algorithms. 

```python
import pandas as pd
import numpy as np
from sklearn import svm,model_selection,preprocessing,metrics
data_dir="/home/eirik/python/AIHack/data/California_metadata/"
```

```python
label_df=pd.read_csv(data_dir+"BG_METADATA_2016.csv")
columnnames={}
for row in label_df.iterrows():
    columnnames[row[1][1]]=row[1][2]
```

```python
#Main dataframe has all variables from this dataset
df=pd.read_csv(data_dir+"X15_EDUCATIONAL_ATTAINMENT.csv")
df.rename(columns=columnnames,inplace=True)
# This dataset contains the variable we want to predict
y_df=pd.read_csv(data_dir+"X19_INCOME.csv")
y_df.rename(columns=columnnames,inplace=True)

y_columnname="PER CAPITA INCOME IN THE PAST 12 MONTHS (IN 2016 INFLATION-ADJUSTED DOLLARS): Total: Total population -- (Estimate)"
#output_df.to_csv("Just for find an interresting variable.csv")
df[y_columnname]=y_df[y_columnname]
# Drop variables used for identification
df.drop("OBJECTID",axis=1,inplace=True)
df.drop("GEOID",axis=1,inplace=True)
# Handle na and split dataset again
df.dropna(inplace=True)
y=df[y_columnname]
df.drop(y_columnname,axis=1,inplace=True)
df.describe()
```

As you can see most block groups actually have a lot of zero values, and all numbers are integers. This is because every single variable in this dataframe represents the number of people fitting their respective variable in each block group.
The next thing we need to do is to normalize each column as per custom. Given the nature of this dataset, where the relations are strong along the rows rather than the columns it might be interesting to try and normalize the values with respect to rows instead.

```python
columns=df.columns.values
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df= pd.DataFrame(x_scaled)
df.head
```

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(df,y,test_size=0.2,random_state=60)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_test, y_test)
```

This is not too bad of a result for simple brute forcing every single value in to the machine, to validate my findings I i did a 10-cross validation

```python
linreg_accuracies = model_selection.cross_val_score(estimator = reg, X = X_train, y = y_train, cv = 10)
print(linreg_accuracies.mean())
print(linreg_accuracies.std())
```

I also wanted to try and see I could improve performance further and there tried a gradient booster aproach. I actually ended up excperimenting quite alot with this algorithm, just to see if could find some nice hyperparameters. Turns out that I almost hit jackpot the first time

```python
import xgboost

xgb0 = xgboost.XGBRegressor(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.75, 
                            colsample_bytree=1, max_depth=7)
xgb0_accuracies = model_selection.cross_val_score(estimator = xgb0 , scoring="r2", X = X_train, y = y_train, cv = 10)
print(xgb0_accuracies.mean())
print(xgb0_accuracies.std())
```

```python
from sklearn.model_selection import GridSearchCV

xgb1=xgboost.XGBRegressor()
xgb1_parameters = [{"n_estimators": [400,500,600], 'learning_rate':[0.04,0.05,0.06],"gamma":[0], "subsample":[0.75],"max_depth":[7],"colsample_bytree":[1]}]
grid_search = GridSearchCV(estimator = xgb1,
                           param_grid = xgb1_parameters,
                           scoring = 'r2',
                           cv = 5,
                           n_jobs = -1)
y_train=np.array(y_train)
grid_search = grid_search.fit(X_train, y_train)
xgb1_best_accuracy = grid_search.best_score_
xgb1_best_parameters = grid_search.best_params_
```

```python
print(xgb1_best_accuracy)
print(xgb1_best_parameters)
```

I did three grid searches for this algorithm:

1. First I tried hyperparameter ={"n_estimators":[50,100,200], "learning_rate":[0.5,0.1,0.05], "gamma":[0,0.1,0.2]}.
    This returned {'gamma': 0, 'learning_rate': 0.1, 'n_estimators': 200}. 

2. From there I tried the variables {"n_estimators":[100,200,400], "learning_rate":[0.06,0.08,0.1], "gamma":[0],...} 
    This returned {'gamma': 0, 'learning_rate': 0.06, 'n_estimators': 400}

3. Lastly I tried {"n_estimators": [400,500,600], 'learning_rate':[0.04,0.05,0.06],"gamma":[0],...} 
    This returned {'colsample_bytree': 1, 'gamma': 0, 'learning_rate': 0.04,...}

Since the score still increases slightly for each iteration, but since this is just a feasibility study there is no reason to keep on tweaking and tuning it.However,there are a lots of more variables available for tuning, we could have changed the max depth of the trees for example to enable catching more advanced realitionships.
We could obviously try and optimize more but for the purposes of this challenge this is enough for now. I also want to try a different gradientboosting regressor, which has a bit different hyperparameters just to see if the performance can be improved.

```python
from sklearn.ensemble import GradientBoostingRegressor

xgb2=GradientBoostingRegressor()
xgb2_parameters = [{"n_estimators": [1000], 'learning_rate':[0.2,0.1,0.05],"alpha":[0.95,0.9,0.8]}]
grid_search = GridSearchCV(estimator = xgb2,
                           param_grid = xgb2_parameters,
                           scoring = 'r2',
                           cv = 5,
                           n_jobs = -1)
y_train=np.array(y_train)
grid_search = grid_search.fit(X_train, y_train)
xgb2_best_accuracy = grid_search.best_score_
xgb2_best_parameters = grid_search.best_params_
```
